### Python has just a few major data types/structures:

- Integers
- Floats
- Strings (Text)
- Lists
- Tuples

### Any datatypes can be stored in variables, created with the `=` sign

```python
sample_int = 23
sample_float = 345.2367
sample_string = "Hello, world!"
sample_list = ["a", 2, 3.5, "c"]
sample_tuple = ("yes", "this")
sample_dict = {"a": 1, "b": 2, "c": 3}

print(type(sample_string))
print(type(sample_int))
```

### Certain operations and functions only work for certain data types

```python
print(sample_int * 2)
print(sample_int + sample_float)
# print(len(sample_int))
print(len(sample_list))
print(len(sample_string))
```

### Lists can:

- contain anything
- be looped through
- be accessed by their index (order)

### Create CSV from lists

```python
print(sample_list)
print() # Empty space makes results easier to read

for l in sample_list: # A for loop to go through each item
    print(l)
print()
    
# A list comprehension does the work of a for loop, with an added conditional "if" statement
new_list = [l for l in sample_list if type(l) == int] 
print(new_list)
print()

print("Second item in list:", sample_list[1])
print("List from first to third item:", sample_list[0:3])
print("List from third item to the end:", sample_list[2:])
```

## Strings are *Lists* of characters!

```python
print(sample_string)
print() # Empty space makes results easier to read

for l in sample_string: # A for loop to go through each item
    print(l)
print()
    
# A list comprehension does the work of a for loop
new_list = [l for l in sample_string] 
print(new_list)
print()

print("Second item in list:", sample_string[1])
print("List from first to third item:", sample_string[0:3])
print("List from third item to the end:", sample_string[2:])
```

## Strings have a bunch of [built-in methods](https://docs.python.org/3/library/stdtypes.html#textseq)

```python
print(sample_string.split()) # Split a string on its spaces
print(sample_string.split(',')) # Split a string on a specific character
print()

print(sample_string.lower()) # convert to lower case
print(sample_string.upper()) # conver to all caps
print()

print(sample_string.strip()) # strip whitesspace around string (In this case, this will do nothing)
print(sample_string.strip('!')) # strip specific character(s)
print(sample_string.strip(',')) # Note that this DOES NOT work
print()

print(sample_string.replace('o','&')) # replace character with another character
print(sample_string.replace(',', '')) # remove a character by "replacing" it with nothing
```

## Accessing and Reading Files!

New key operators/functions:

- with
- as
- open
- read

```python
with open('data/fq1.txt', 'r') as newfile:
    print("File without using 'read':", newfile)
    print()
    
    text = newfile.read() # Store this in a variable or you won't be able to use it later
    print(text)
```

```python
# Split texts by lines
lines = text.split("\n")
print(lines)
```

```python
# Get rid of whitespace
cleaned_lines = [l.strip() for l in lines]
print(cleaned_lines)
```

```python
# Split text by space
words = text.split(" ")
print(words)
```

## But wait! What about those newlines?

For that, we need to *import* a special [*module* for regular expressions](https://docs.python.org/3.7/library/re.html).

```python
import re # Import regular expressions module

better_words = re.split("\s+", text) # Split on all whitespace characters
print(better_words)
```

```python
better_words_no_punct = re.split('\W+', text) # Split on all non-word characters
print(better_words_no_punct)
```
